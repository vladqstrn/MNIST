```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Vlad/OneDrive/Документы/fashion/fashion-mnist_train.csv')
```

```python
CLASS_NUM = df['label'].value_counts().count()
WIDTH = 28
HEIGHT = 28
```

```python
labels = {0 : 'T-shirt/top', 1: 'Trouser', 2: 'Pullover', 3: 'Dress', 4: 'Coat',
          5: 'Sandal', 6: 'Shirt', 7: 'Sneaker', 8: 'Bag', 9: 'Ankle Boot'}
```

```python
class Image_show():
    def __init__(self, df, labels, WIDTH, HEIGHT):
        self.df = df
        self.labels = labels
        self.WIDTH = WIDTH
        self.HEIGHT = HEIGHT
    
    def _sample(self):
        images = []
        labels = []
        for key in self.labels.keys():
            sample = df[df['label'] == key].head(2)
            for index, _ in enumerate(sample.values):
                img = np.array(sample.iloc[index, 1:]).reshape(self.WIDTH, self.HEIGHT)
                images.append(img)
                labels.append(sample.iloc[index, 0])
        return images, labels
    
    def _print_samples(self, images, d_labels):
        _, ax = plt.subplots(10,2, figsize=(5,20))

        for index, img in enumerate(images):
            ax[index//2, index%2].imshow(img)
            ax[index//2, index%2].axis('off')
            ax[index//2, index%2].set_title(labels[d_labels[index]])
        plt.show()
    
    def show(self):
        images, labels = self._sample()
        self._print_samples(images, labels)
```

```python
image = Image_show(df, labels, WIDTH, HEIGHT)
image.show()
```

```python
import tensorflow as ts
```

```python
def process_df(df):
    labels = ts.keras.utils.to_categorical(df.label, CLASS_NUM)
    num_images = df.shape[0]
    x = df.values[:,1:]
    x = x.reshape(num_images, WIDTH, HEIGHT, 1)
    images = x / 255
    return images, labels
```

```python
train, target = process_df(df)
```

```python
from tensorflow.python.keras.models import Sequential
from keras import layers
import keras
```

```python
model = Sequential([
    layers.Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 kernel_initializer='he_normal',
                 input_shape=(WIDTH, HEIGHT, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(CLASS_NUM, activation='softmax')]
)

```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer='adam',
              metrics=['accuracy'])
```

```python
train_model = model.fit(train, target,
                  batch_size=64,
                  epochs=1,
                  verbose=1)
                  #validation_data=(X_val, y_val))
```
